```python
import json
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_accuracy_from_jsonl(metrics_file_path: str, use_log_scale: bool = False):
    steps = []
    train_accs = []
    val_accs = []

    # Read the JSONL file
    with open(metrics_file_path, 'r') as f:
        for line in f:
            data = json.loads(line.strip())
            steps.append(data['step'])
            train_accs.append(data['train_acc'])
            val_accs.append(data['val_acc'])

    # Create the plot
    plt.figure(figsize=(6, 4))

    # Plot both train and validation accuracy
    # plt.semilogx(steps, train_accs, 'b-', label='Train Accuracy', linewidth=2)
    # plt.semilogx(steps, val_accs, 'r-', label='Validation Accuracy', linewidth=2)
    #
    # plt.xlabel('Step (log scale)')

    # Plot both train and validation accuracy
    if use_log_scale:
        plt.semilogx(steps, train_accs, 'b-', label='Train Accuracy', linewidth=2)
        plt.semilogx(steps, val_accs, 'r-', label='Validation Accuracy', linewidth=2)
        plt.xlabel('Step (log scale)')
    else:
        plt.plot(steps, train_accs, 'b-', label='Train Accuracy', linewidth=2)
        plt.plot(steps, val_accs, 'r-', label='Validation Accuracy', linewidth=2)
        plt.xlabel('Step')

    plt.ylabel('Accuracy')
    plt.title('Training and Validation Accuracy vs Step')
    plt.legend()
    plt.grid(True, alpha=0.3)

    # Set y-axis to be more appropriate for accuracy (0 to 1)
    plt.ylim(0, 1.05)

    plt.tight_layout()
    plt.show()
```

```python
file_path = "/home/taot/github/grokking-lab/runs/train_frac/20260121_082404/metrics.jsonl"
plot_accuracy_from_jsonl(file_path, use_log_scale=False)
```
